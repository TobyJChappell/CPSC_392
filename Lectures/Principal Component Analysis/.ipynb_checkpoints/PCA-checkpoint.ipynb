```python
import pandas as pd
from plotnine import *
songs = pd.read_csv("songs.csv")
songs.head()
```

```python
ggplot(songs,aes(x="tempo",y="energy",color="genre"))+geom_point()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=4)
km.fit(songs)
songs["labels"] = km.labels_
songs.head()
```

```python
ggplot(songs,aes(x="tempo",y="energy",color="factor(labels)"))+geom_point()
```

```python
# scale values
songs = pd.read_csv("songs.csv")
from sklearn.preprocessing import StandardScaler
songs = StandardScaler().fit_transform(songs)
songs
```

```python
# normalize values
from sklearn.preprocessing import Normalizer
songs = Normalizer().fit_transform(songs)
songs
```

```python
#reduce 10 features into 2
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(songs)
principalComponents
```

```python
songs = pd.read_csv("songs.csv")
songs["PC1"] = principalComponents[:,0]
songs["PC2"] = principalComponents[:,1]
```

```python
ggplot(songs,aes(x="PC1",y="PC2",color="genre"))+geom_point()
```

```python
pca.explained_variance_ratio_
```
