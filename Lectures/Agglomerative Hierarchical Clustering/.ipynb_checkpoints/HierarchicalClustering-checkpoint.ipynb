```python
import pandas as pd
songs = pd.read_csv("songs.csv")
songs.head()
```

```python
features = songs.iloc[:,0:10]
features
```

```python
from sklearn.preprocessing import StandardScaler
features = StandardScaler().fit_transform(features)
features
```

```python
from sklearn.preprocessing import Normalizer
features = Normalizer().fit_transform(features)
```

```python
features
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(features)
```

```python
songs["PC1"] = principalComponents[:,0]
songs["PC2"] = principalComponents[:,1]
songs.head()
```

```python
X = songs[["PC1","PC2"]]
X
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linkage_matrix = linkage(X,"ward")
figure = plt.figure(figsize=(7.5,5))

dendrogram(linkage_matrix)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("cluster")
plt.ylabel("distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(linkage="ward",n_clusters=3)
clustering.fit(X)
```

```python
songs["labels"] = clustering.labels_
songs.head()
```

```python
songs.groupby(["labels"]).size()
```

```python
from plotnine import *
ggplot(songs,aes(x="PC1",y="PC2",color="genre"))+geom_point()
```

```python
ggplot(songs,aes(x="PC1",y="PC2",color="factor(labels)"))+geom_point()
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(linkage="ward",n_clusters=4)
clustering.fit(X)
songs["labels"] = clustering.labels_
ggplot(songs,aes(x="PC1",y="PC2",color="factor(labels)"))+geom_point()
```
